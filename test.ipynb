```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt


model = YOLO("C:/Users/user/Desktop/test/ultralytics-cpu/runs/detect/train3/weights/best.pt")


image_path = r"C:\Users\user\Desktop\test\ultralytics-cpu\ultralytics\data\dataset\african-american-man-wearing-mask-600nw-2601221721.jpg"


img = cv2.imread(image_path)
if img is None:
    print("❌ Could not read the image. Please check the path.")
else:
    results = model.predict(source=img, conf=0.25, verbose=False)
    annotated_img = results[0].plot()

   
    rgb_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

   
    plt.figure(figsize=(10, 8))
    plt.imshow(rgb_img)
    plt.axis('off')
    plt.title("🧠 YOLOv8 Detection")
    plt.show()
```
